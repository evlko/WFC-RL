```python
from project.machine_learning.dataset import Dataset
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer
import matplotlib.pyplot as plt
```

```python
wfc_df = Dataset.load_data("data/graphs/wfc")
wfcwmc_df = Dataset.load_data("data/graphs/wfcwmc")
```

```python
model = KElbowVisualizer(KMeans(n_init=10, random_state=42), k=20, timings=False, title=" ")
model.fit(wfc_df)
model.show(outpath="content/wfc_elbow_plot.png",  bbox_inches="tight")
```

```python
model = KElbowVisualizer(KMeans(n_init=10, random_state=42), k=20, timings=False, title=" ")
model.fit(wfcwmc_df)
model.show(outpath="content/wfcwmc_elbow_plot.png",  bbox_inches="tight")
```
